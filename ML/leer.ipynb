####  Realizar la carga de datos desde Python (Google Colab) o R-studio. 



```python
#importo la libreria y le pongo el alias de "pd"
import pandas as pd

# Creo una lista que contiene el nombre de las columnas de la base de datos
COLUMNS = ["age","workclass","fnlwgt","education","education_num","marital","occupation",
           "relationship","race","sex","capital_gain","capital_loss","hours_week","native_country","label"]
url ="https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

#creo un try except por si falla la lectura de la base de datos, no se pare el programa
try:
#Leo la base de datos desde la url
    data_set = pd.read_csv(url,names=COLUMNS)
except:
    print("No se encontro el URL")

#Imprimo una muestra de la base de datos desde arriba(head) y desde abajo(tail)

print(data_set.head())
print(data_set.tail())

#Imprimo la cantidad de filas y columnas que tiene la base de datos
print("Data set con {} Filas y {} Columnas".format(data_set.shape[0],data_set.shape[1]))

#Imprimo nformaci칩n general de la base de datos
data_set.info()

```

####   An치lisis descriptivo de los datos que incluya las medidas de tendencia central.
Se Analiza el data set. Identificando, de acuerdo a las columnas, es un dataset que describe individuos con variables como edad, nivel educativo, estado civil. etc
1) describe
2) describe all, para las columnas que no se analizaba con el describe
3) Se agruparon los valores de cada columna, para ver los registros que m치s y que menos se repiten
4) Se analiz칩 las medidas de tendencia centrla



```python
data_set.describe()
# Como no me describe todos los campos, fuerzo al describe a que describa todos los campos
data_set.describe(include='all')
# Ciclo qe recorre cada columna, agrupa los distintos valores y muestra los que m치s se repiten hasta los que menos se repiten
for columna in data_set.columns:
    resultado = data_set.groupby(columna).size().reset_index(name='registros')
    resultado = resultado.sort_values(by='registros', ascending=False,ignore_index=True)
    print("*"*30)
    print(columna)
    print("*"*30)
    print(resultado)

#Codigo que recorre cada columna y obtiene 
    #1) media
    #2) mediana
    #3) Desviaci칩n estandar
    #4) Moda
for columna in data_set.columns:
    try:
        print("*"*30)
        print(columna)
        print("*"*30)
        print("La media de {} es = {}".format(columna, data_set[columna].mean()))
        print("La mediana de {} es = {}".format(columna, data_set[columna].median()))
        print("La Desviacion estandar de {} es = {}".format(columna, data_set[columna].std()))
        print("Las modas de {} son  = ".format(columna),end="")
        modas =[]
        for moda in data_set[columna].mode():    
            modas.append(moda)   
        print(modas)    
    except:
        print("No se pudo obtener la medida de tendencia central para la columna "+ columna)
    finally:
        print("\n"*2)
```

#### Tratamiento de missing.


```python
# Del data set suma todos los nulos por columna
data_set.isnull().sum()
```


#### Identificaci칩n y eliminaci칩n de errores.


```python
# No hay ningun valor en nulo. Sin embargo, existe un caracter raro " ? " analizaremos los registros con estos valores,
# este caracter tiene espacios en blanco por lo cual se le elimina espacio en blanco con el metodo strip

#Nuevo data_set con el valor de occupation filtrado 
data_set_2 = data_set[data_set['occupation'].str.strip()!= "?"]
for columna in data_set_2.columns:
    resultado = data_set_2.groupby(columna).size().reset_index(name='registros')
    resultado = resultado.sort_values(by='registros', ascending=False,ignore_index=True)
    print("*"*30)
    print(columna)
    print("*"*30)
    print(resultado)
```
